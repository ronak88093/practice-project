# DATA PREDICTION

```python
x=data.drop(['charges'],axis=1) # independent variable
y=data['charges']               # dependent variable
```

```python
x
```

```python
y
```

```python
# dividing the dataset in training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
x.shape,y.shape
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# importing model
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report 
```

**TRAINING THE MODEL**

```python
lr= LinearRegression()
lr.fit(x_train, y_train)

rf=RandomForestRegressor()
rf.fit(x_train, y_train)

gb=GradientBoostingRegressor()
gb.fit(x_train, y_train)


dtr=DecisionTreeRegressor()
dtr.fit(x_train, y_train)

svr=SVR()
svr.fit(x_train, y_train)
```

**CLACULATE ACCURACY**

```python
acc=lr.score(x_test,y_test)
print("accuracy score of LR is = ",acc*100,"%")
acc=rf.score(x_test,y_test)
print("accuracy score RF is = ",acc*100,"%")
acc=gb.score(x_test,y_test)
print("accuracy score GB is = ",acc*100,"%")
acc=dtr.score(x_test,y_test)
print("accuracy score DTR is = ",acc*100,"%")
acc=svr.score(x_test,y_test)
print("accuracy score SVR is = ",acc*100,"%")
```

**BEST MODEL FOR TESTING IS Gradient Boosting Regressor**

# hyper perameter tuning for GradientBoostingRegressor

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2, 3]
}
```

```python
grid1 = GridSearchCV(estimator=gb, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
best_regressor = grid1.best_estimator_
best_regressor
```

```python
final_gb = GradientBoostingRegressor(learning_rate=0.01, n_estimators=50)

final_gb.fit(x_train,y_train)
predgb=final_gb.predict(x_test)
acc=gb.score(x_test,y_test)
print("accuracy score GB is = ",acc*100,"%")
```
